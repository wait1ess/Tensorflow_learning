```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
# 载入数据
mnist = input_data.read_data_sets('D:/kaggle-10-most-popular-dataset/MNIST_data/', one_hot=True)
```

```python
# 批次的大小
batch_size = 100
n_batch = mnist.train.num_examples // batch_size

x = tf.placeholder(tf.float32, [None,784])
y = tf.placeholder(tf.float32, [None, 10])
keep_prob = tf.placeholder(tf.float32)

# 创建神经网络
                                                              # 输入层 784个神经元
W1 = tf.Variable(tf.truncated_normal([784,2000],stddev=0.1))  # 利用截断的正态分布进行权值初始化
b1 = tf.Variable(tf.zeros([1, 2000])+0.1)
                                                              # 激活层
layer1 = tf.nn.relu(tf.matmul(x,W1) + b1)
                                                              # drop层 keep_prob表示该层百分多少神经元在工作
layer1_drop = tf.nn.dropout(layer1,keep_prob=keep_prob)

                                                              # 第二层隐藏层 2000个神经元
W2 = tf.Variable(tf.truncated_normal([2000,500],stddev=0.1))
b2 = tf.Variable(tf.zeros([1, 500]))
layer2 = tf.nn.relu(tf.matmul(layer1_drop,W2) + b2)
layer2_drop = tf.nn.dropout(layer2,keep_prob=keep_prob)

                                                              # 第三层隐藏层 500个神经元
W3 = tf.Variable(tf.truncated_normal([500,10],stddev=0.1))
b3 = tf.Variable(tf.zeros([1,10]))
                                                             # 输出层 10个神经元
prediction = tf.nn.sigmoid(tf.matmul(layer2_drop,W3) + b3)
                                                             # 定义损失函数 交叉熵代价函数
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=y, logits=prediction))

                                                             # 优化方法为  梯度下降法
# train_step = tf.train.GradientDescentOptimizer(0.2).minimize(loss)#得到97的正确率
train_step = tf.train.AdadeltaOptimizer(0.1).minimize(loss)


# 初始化变量
init = tf.global_variables_initializer()

prediction_2 = tf.nn.softmax(prediction)
# 得到一个布尔型列表，存放结果是否正确
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(prediction_2,1)) #argmax 返回一维张量中最大值索引

# 求准确率
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))        # 把布尔值转换为浮点型求平均数

with tf.Session() as sess:
    sess.run(init)
    for epoch in range(100):
        for batch in range(n_batch):
                                                                  # 获得批次数据
            batch_xs, batch_ys = mnist.train.next_batch(batch_size)
            sess.run(train_step, feed_dict={x:batch_xs, y:batch_ys, keep_prob:0.8})                    # 训练   (一般优化算法为最终OP， 往最初的占位符喂数据)
        acc = sess.run(accuracy, feed_dict={x:mnist.test.images,y:mnist.test.labels,keep_prob:1.0} )   # 测试
        print("Iter " + str(epoch) + " Testing Accuracy: " + str(acc))
```
